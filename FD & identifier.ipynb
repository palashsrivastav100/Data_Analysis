```python
def face_detect(img):

  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  result = hog_face_detector(img_gray)

  for bbox in result:
  # print(bbox.left())
  # print(bbox.right())
  # print(bbox.top())
  # print(bbox.bottom())
    left = bbox.left()
    top = bbox.top()
    right = bbox.right()
    bottom = bbox.bottom()
    cv2.rectangle(img,(left, top),(right, bottom),(255,0,0),15)




  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img_rgb)
```

```python
def face_mapping(img_add):

  hog_face_detector = dlib.get_frontal_face_detector()
  facial_map = dlib.shape_predictor(r"C:\Users\palash\Downloads\shape_predictor_68_face_landmarks.dat")

  img = cv2.imread(img_add)
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  results = hog_face_detector(gray)
  for result in results:
    landmarks = facial_map(gray, result)
    for i in range(68):
      x = landmarks.part(i).x
      y = landmarks.part(i).y
      img = cv2.circle(img, (x,y), 5, [0,255,255], -1)

  plt.imshow(img[:,:,::-1])

```

```python
face_mapping(r"C:\Users\palash\Downloads\HR_UKG.jpg")
```
