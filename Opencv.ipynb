```python
import cv2
import numpy as np

img = np.zeros((512, 512, 3), np.uint8) #uint8: 0 to 255

plt.imshow(img)
```

```python
#BGR Image . It is represented in Blue, Green and Red Channels...
image = cv2.imread(r"C:\Users\palash\Downloads\7-colorful-photography.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
lower_b = np.array([200,200,200])
upper_b = np.array([255,255,255])

mask = cv2.inRange(img, lower_b, upper_b) # OpenCV function
plt.imshow(mask)
```

```python
hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```

```python
def inRange(img, lower_b, upper_b):
    ch1, ch2, ch3 = cv2.split(img)
    ch1m = (lower_b[0] <= ch1) & (ch1 <= upper_b[0])
    ch2m = (lower_b[1] <= ch2) & (ch2 <= upper_b[1])b
    ch3m = (lower_b[2] <= ch3) & (ch3 <= upper_b[2])
    mask = ch1m & ch2m & ch3m
    return mask.astype(np.uint8)*255
```

```python
hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
# Blue Color Triangle
lower_hue = np.array([35,0,0])
upper_hue = np.array([65, 255,255])
```

```python
mask = cv2.inRange(hsv,lower_hue,upper_hue)
# cv2_imshow(mask)
plt.imshow(mask, cmap = "gray")
result = cv2.bitwise_and(image, image, mask = mask)
result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
plt.imshow(result_rgb)
```
